```python
import pandas as pd
from tensorflow.keras import *
from sklearn.preprocessing import StandardScaler
import copy

# Read CSV
fram_train = pd.read_csv("Framingham_Training.csv")

# Split/Modify Data
Age     = fram_train["Age"]

# Normalization
Age_max = max(Age)
Age_min = min(Age)
tmp     = (Age - Age_min)/(Age_max - Age_min)

# Append Normirized Age to DataFrame
fram_train = fram_train.assign(Age_mm=tmp)

# Define Input and Output for Training
x = fram_train[["Sex","Age_mm"]]
y = fram_train["Death"]

# Define Neural Network Architecture Model
model = models.Sequential()
model.add(layers.Dense(1, activation="sigmoid"))
model.add(layers.Dense(1, activation="sigmoid"))

# Compile model
model.compile(optimizer='adam', loss='mean_absolute_error', metrics=['acc'])

# Setting for training
model.fit(x, y, epochs=40)

# Output Weights and Bias of Dense Layer
weight_1st = model.layers[0].get_weights()[0]
bias_1st   = model.layers[0].get_weights()[1]
weight_2nd = model.layers[1].get_weights()[0]
bias_2nd   = model.layers[1].get_weights()[1]

table = pd.DataFrame([["W_B1H", "W_I1H1", "W_I2H1", "W_B2O1", "W_H1O1"],
                      [bias_1st, weight_1st[0], weight_1st[1], bias_2nd, weight_2nd[0]]
                     ])
table
```
