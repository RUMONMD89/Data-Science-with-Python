```python
# Library import
from pandas.core.apply import NDFrameApply
from os import fpathconf
#from pandas.core.resample import f
import pandas as pd
import numpy  as np
import statsmodels.tools.tools as stattools
from sklearn.tree    import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import confusion_matrix

# Read data
clothing_train = pd.read_csv("clothing_data_driven_training.csv")
clothing_test = pd.read_csv("clothing_data_driven_test.csv")

# Training data
clothing_train_Response = clothing_train["Response"]
y = clothing_train_Response
y_names = ["0", "1"]

x = clothing_train.drop(columns=["Response"])
x_names = ["Days.since.Purchase","Number.of.Purchase.Visits","Sales.per.Visit"]

# Test data
test_x = clothing_test.drop(columns=["Response"])
actual = clothing_test["Response"]

# Define costs and train
c5 = DecisionTreeClassifier(criterion="entropy", max_leaf_nodes=5, class_weight=None).fit(x, y)
export_graphviz(c5, out_file="c5.dot", feature_names=x_names, class_names=y_names)

# Prediction
ypred = c5.predict(test_x)

# Visialization
cm = confusion_matrix(actual,ypred)
t1 = pd.DataFrame([
                   [" ", "Predicted: 0","Predicted: 1", "Total"],
                   ["Actual: 0", cm[0][0], cm[0][1], sum(cm[0])],
                   ["Actual: 1", cm[1][0], cm[1][1], sum(cm[1])],
                   ["Total", cm[0][0]+cm[1][0], cm[0][1]+cm[1][1], sum(sum(cm))]
                  ])
t1
```

```python
# Define costs for each column and train
weights  = {0:1, 1:4}
c5_costs = DecisionTreeClassifier(criterion="entropy", max_leaf_nodes=5, class_weight=weights).fit(x, y)
export_graphviz(c5_costs, out_file="c5_costs.dot", feature_names=x_names, class_names=y_names)

# Prediction
ypred_costs = c5_costs.predict(test_x)

cm = confusion_matrix(actual,ypred_costs)

t1 = pd.DataFrame([
                   [" ", "Predicted: 0","Predicted: 1", "Total"],
                   ["Actual: 0", cm[0][0], cm[0][1], sum(cm[0])],
                   ["Actual: 1", cm[1][0], cm[1][1], sum(cm[1])],
                   ["Total", cm[0][0]+cm[1][0], cm[0][1]+cm[1][1], sum(sum(cm))]
                  ])
t1
```
